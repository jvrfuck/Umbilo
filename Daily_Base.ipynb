# Simulated data of Q varying in situation 1

Q = flow

TN = total nitrogen

TKN = total nitrogen Kjeldahl

NH = ammoniacal nitrogen

TP = total phosphorus

PO = orthophosphate

```python
df_entrada = pd.read_excel('ETE Umbilo - Dados de entrada - escala diária.xlsx')

df_saida = pd.read_excel('ETE Umbilo - Dados de saída simulados - escala diária.xlsx')


df_Umbilo = pd.merge(left=df_entrada,right=df_saida, right_on='.t',left_on='.t')
df_Umbilo
```

```python
df_Umbilo.columns
```

```python
df_Umbilo.columns = ['Dia', 'Q_entrada','NH_entrada','TSS_entrada','DQO_entrada','DBO_entrada',
                         'TN_entrada','TKN_entrada','TP_entrada','PO_entrada',
                         'Q_saida','NH_saida','NO_saida','TSS_saida','DQO_saida',
                         'DBO_saida','TN_saida','TKN_saida','TP_saida','PO_saida']
df_Umbilo.set_index('Dia',inplace=True)

df_Umbilo.to_csv('Base_Umbilo_Diaria.csv',sep=';',decimal=',',encoding='utf-8-sig')

df_Umbilo
```
