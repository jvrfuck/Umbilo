```python
import numpy as np
import pandas as pd
```

```python
df_saida = pd.read_csv('ETE Umbilo - Dados de saída medidos.txt', sep='\t')

df_saida.columns = df_saida.iloc[1]
df_saida = df_saida.iloc[4:,:]
# df_saida.index = df_saida['t']
# df_saida = df_saida.drop('t', axis=1)

df_saida = df_saida.astype(float)
df_saida.isna().sum()  # dia 307 missing Nitrate

df_saida
```

```python
df_entrada = pd.read_excel('ETE Umbilo - Dados de entrada - escala diária.xlsx')

df_real = pd.merge(left=df_entrada,right=df_saida, right_on='t',left_on='.t')
df_real.isna().sum()

df_real = df_real.drop('t',axis=1)

df_real
```

```python
df_real.columns = ['Dia', 'Q_entrada','NH_entrada','TSS_entrada','DQO_entrada','DBO_entrada',
                         'TN_entrada','TKN_entrada','TP_entrada','PO_entrada',
                         'DQO_saida','Nitrato_saida','Amonia_saida']

df_real = df_real.drop(['PO_entrada','TKN_entrada'], axis=1)
df_real.set_index('Dia',inplace=True)

df_real.to_csv('Real_Umbilo.csv',sep=';',decimal=',',encoding='utf-8-sig')

df_real
```
